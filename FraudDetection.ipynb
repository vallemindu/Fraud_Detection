```python
import pandas as pd

df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['Time', 'Class'], axis=1)
y = df['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression


lr_model = LogisticRegression(random_state=0, max_iter=5000)
lr_model.fit(x_train, y_train)
```

```python
import matplotlib
matplotlib.use('TkAgg')
```

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

predictions = lr_model.predict(x_test)

labels = ['Legitimate', 'Fraudulent']

cm = confusion_matrix(y_test, predictions, labels=lr_model.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,

                              display_labels=lr_model.classes_)

disp.plot()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=0)
rf_model.fit(x_train, y_train)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

predictions = rf_model.predict(x_test)

labels = ['Legitimate', 'Fraudulent']

cm = confusion_matrix(y_test, predictions, labels=rf_model.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,

                              display_labels=rf_model.classes_)

disp.plot()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_model = GradientBoostingClassifier(random_state=0)
gbm_model.fit(x_train, y_train)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

predictions = gbm_model.predict(x_test)

labels = ['Legitimate', 'Fraudulent']

cm = confusion_matrix(y_test, predictions, labels=gbm_model.classes_)

dis = ConfusionMatrixDisplay(confusion_matrix=cm,

                              display_labels=gbm_model.classes_)

dis.plot()
```
